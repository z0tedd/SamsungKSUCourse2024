<h1>1. История развития компьютерного зрения</h1>

<h2>Ранние исследования</h2>

<p><strong>История развития компьютерного зрения</strong> берет свое начало в середине 20-го века, когда ученые начали задумываться о том, как машины могут воспринимать и интерпретировать визуальную информацию так же, как это делает человеческий мозг. Одной из ключевых фигур в становлении этой области стал <em>Дэвид Марр</em>, британский нейробиолог и теоретик, чья работа заложила фундамент для современных исследований в области визуального восприятия и компьютерного зрения.</p>

<p>Дэвид Марр начал свою карьеру в области нейробиологии, изучая, как нейронные сети в мозге обрабатывают информацию. Его интерес к визуальному восприятию привел его к разработке теоретической рамки, которая объединяла нейробиологические данные с вычислительными моделями. В своей книге "<em>Vision</em>", опубликованной в 1982 году, Марр представил целостный подход к пониманию зрительной системы, предложив многоуровневую модель анализа визуальной информации.</p>

<p>Марр утверждал, что для полного понимания любой сложной информационной системы необходимо рассмотреть ее на <strong>трех уровнях</strong>: вычислительном, алгоритмическом и уровне реализации. Вычислительный уровень отвечает на вопрос "что" система делает и "почему" она это делает. Алгоритмический уровень фокусируется на том, какие алгоритмы и представления данных используются для выполнения задач. Уровень реализации касается того, как эти алгоритмы воплощены в физической системе, будь то биологический мозг или электронное устройство.</p>

<p>Применяя эту трехуровневую модель к зрительной системе, Марр предложил, что зрение можно рассматривать как процесс преобразования информации из двумерного изображения на сетчатке в трехмерное представление окружающего мира. На вычислительном уровне задача зрения заключается в восстановлении структурных свойств объектов из визуальных данных. На алгоритмическом уровне это включает в себя разработку алгоритмов для обнаружения краев, контуров, текстур и других визуальных признаков. На уровне реализации рассматривается, как эти алгоритмы могут быть воплощены в нейронных сетях или электронных схемах.</p>

<p>Одним из ключевых вкладов Марра было понимание того, что зрение не является пассивным процессом, а активно строит внутренние представления на основе входной информации. Он предложил концепцию "<em>сырых эскизов</em>" (raw primal sketch), где зрительная система сначала выделяет основные визуальные признаки, такие как изменения яркости и контраста. Затем эти признаки объединяются в более сложные структуры на последующих этапах обработки, что приводит к формированию трехмерной модели окружающего мира.</p>

<p>Марр также подчеркнул важность интеграции данных из различных источников и масштабов. Он рассматривал зрительную систему как иерархическую, где обработка информации происходит на разных уровнях абстракции. Это позволило объяснить, как мы можем распознавать объекты независимо от их положения, масштаба или ориентации в пространстве.</p>

<p>Работа Марра имела глубокое влияние на развитие компьютерного зрения и нейронауки. Его теоретические модели предоставили основу для разработки алгоритмов обработки изображений, которые используются в современных системах распознавания образов, робототехнике и других областях. Подход Марра к разделению сложных проблем на более простые подзадачи и рассмотрению их на разных уровнях анализа стал стандартом в области исследований искусственного интеллекта.</p>

<p>Кроме того, Марр подчеркивал важность <strong>междисциплинарного подхода</strong>. Он верил, что понимание визуального восприятия требует объединения знаний из нейробиологии, психологии и компьютерных наук. Это видение стимулировало сотрудничество между различными научными дисциплинами и способствовало прогрессу в понимании того, как машины и биологические системы могут обрабатывать визуальную информацию.</p>

<p>Одним из примеров практического применения идей Марра является разработка алгоритмов обнаружения краев в изображениях. Эти алгоритмы основаны на понимании того, что изменения яркости и контраста являются ключевыми признаками для идентификации границ объектов в визуальной сцене. Такие методы используются в цифровых камерах, системах навигации для автономных транспортных средств и многих других приложениях.</p>

<p>Еще одним значимым вкладом Марра было его исследование стереовидения. Он изучал, как зрительная система использует небольшие различия в изображениях, полученных каждым глазом, для оценки глубины и трехмерной структуры объектов. Это привело к развитию алгоритмов стереоматчинга, которые позволяют реконструировать трехмерные сцены из двухмерных изображений. Такие технологии нашли применение в виртуальной реальности, 3D-сканировании и робототехнике.</p>

<p>Марр также обращал внимание на проблемы движения и его восприятия. Он исследовал, как последовательные изображения могут использоваться для определения направления и скорости движения объектов. Это направление исследований стало основой для развития методов оптического потока, которые используются для отслеживания объектов в видео, стабилизации изображения и других динамических приложений.</p>

<p>Важно отметить, что подход Марра был революционным для своего времени. В то время многие исследования были сосредоточены либо на чисто биологических аспектах зрения, либо на разработке технических решений без учета биологических принципов. Марр объединил эти области, показав, что глубокое понимание биологических систем может вдохновить эффективные вычислительные модели, и наоборот.</p>

<p>После публикации "<em>Vision</em>" многие исследователи продолжили развивать идеи Марра, применяя его принципы к новым задачам и технологиям. Его работа стимулировала развитие конволюционных нейронных сетей, которые сегодня являются основой для многих приложений в области глубокого обучения и компьютерного зрения. Хотя Марр не дожил до того времени, когда эти технологии стали широко распространенными, его влияние ощущается в каждом аспекте современных исследований в этой области.</p>

<p>В заключение, вклад Дэвида Марра в развитие компьютерного зрения и теории визуального восприятия трудно переоценить. Его многоуровневый подход к анализу сложных систем предоставил исследователям мощный инструмент для понимания и моделирования зрительных процессов. Идеи Марра продолжают вдохновлять новые поколения ученых и инженеров, работающих над тем, чтобы машины могли видеть и понимать мир так же эффективно, как это делает человеческий мозг.</p>

<p>История развития компьютерного зрения неразрывно связана с работой таких пионеров, как Марр, чьи междисциплинарные исследования заложили основу для современных технологий. Сегодня, благодаря его теоретическим достижениям, мы имеем системы искусственного интеллекта, способные распознавать лица, анализировать медицинские изображения, управлять автономными транспортными средствами и выполнять множество других задач, которые ранее казались невозможными.</p>

<h2>Развитие методов</h2>

<p><strong>Развитие алгоритмов и методов в области компьютерного зрения</strong> прошло долгий путь от простых операций обработки изображений до современных методов машинного обучения. В начале своего становления компьютерное зрение опиралось на фундаментальные математические и физические принципы, позволяющие компьютерам обрабатывать и анализировать цифровые изображения. Первые шаги в этой области были связаны с применением базовых алгоритмов для решения простых задач, таких как улучшение качества изображения, удаление шума и выделение контуров.</p>

<p>Одним из ключевых направлений ранних исследований было обнаружение и выделение краев на изображениях. Алгоритмы, такие как <em>оператор Собеля</em>, Прюитта и метод Кэнни, стали основой для выделения существенных особенностей в визуальных данных. Эти методы использовали градиенты яркости для определения мест, где происходят резкие изменения интенсивности, что соответствовало границам объектов в сцене. Несмотря на свою простоту, такие алгоритмы оказались эффективными инструментами для предварительной обработки изображений и закладывали фундамент для более сложных задач распознавания.</p>

<p>Параллельно с развитием методов выделения краев происходило совершенствование техник фильтрации и преобразования изображений. Применение сверток, фильтров Гаусса и других математических инструментов позволило улучшить качество изображений и подготовить их для дальнейшего анализа. Преобразование Фурье и его дискретная версия стали важными инструментами для анализа частотных составляющих изображения, что нашло применение в сжатии данных и распознавании текстур.</p>

<p>С развитием вычислительной техники и увеличением мощности компьютеров появилась возможность разрабатывать более сложные алгоритмы, способные не только обрабатывать, но и интерпретировать визуальные данные. Одним из важных этапов стало создание методов обнаружения и описания особенностей на изображениях. Алгоритмы, такие как <em>SIFT</em> (Scale-Invariant Feature Transform) и <em>SURF</em> (Speeded Up Robust Features), позволили выделять ключевые точки и создавать их описания, устойчивые к изменениям масштаба, вращения и освещения. Это открыло новые возможности для задач сопоставления изображений, стежки панорам и 3D-реконструкции.</p>

<p>В этот период также начали применяться статистические методы и вероятностные модели для решения задач компьютерного зрения. Использование методов кластеризации, анализа главных компонент и байесовских моделей позволило обрабатывать большие объемы данных и извлекать из них значимые паттерны. Такие подходы были особенно полезны для распознавания объектов и классификации изображений, где необходимо было учитывать разнообразие и вариативность реальных данных.</p>

<p>Однако с усложнением задач стало очевидно, что традиционные алгоритмические подходы имеют свои ограничения. Для решения более сложных проблем, таких как распознавание лиц, эмоций или жестов, требовались новые методы, способные обучаться на данных и адаптироваться к новым ситуациям. Это привело к интеграции методов машинного обучения в область компьютерного зрения.</p>

<p>Первые попытки применения машинного обучения в компьютерном зрении были связаны с использованием классических алгоритмов, таких как метод опорных векторов (<em>SVM</em>), деревья решений и <em>$k$-ближайших соседей</em>. Эти методы позволяли строить модели, способные классифицировать объекты на изображениях на основе извлеченных признаков. Однако эффективность таких подходов сильно зависела от качества и релевантности вручную разработанных признаков, что ограничивало их применение.</p>

<p>Настоящим прорывом стало возрождение интереса к нейронным сетям и появление глубокого обучения. Хотя нейронные сети были известны еще с середины 20 века, их широкое применение стало возможным только с развитием вычислительных мощностей и появлением больших объемов данных. В 2012 году на соревновании <em>ImageNet</em> произошел значимый прорыв: модель <em>AlexNet</em>, основанная на глубоких сверточных нейронных сетях (<em>CNN</em>), значительно превзошла все предыдущие результаты в задаче классификации изображений. Это событие ознаменовало новую эру в компьютерном зрении и стимулировало интенсивные исследования в области глубинного обучения.</p>

<p>Сверточные нейронные сети оказались особенно эффективными для обработки визуальных данных благодаря своей архитектуре, которая учитывает пространственную структуру изображений. Использование сверток, пулинга и нелинейных активаций позволило моделям автоматически извлекать сложные иерархические признаки из данных, устраняя необходимость в ручной разработке признаков. Это привело к быстрому прогрессу в различных задачах, включая распознавание объектов, сегментацию изображений и детектирование объектов.</p>

<p>После успеха AlexNet последовали более глубокие и сложные модели, такие как <em>VGGNet</em>, <em>GoogLeNet</em> и <em>ResNet</em>. Каждая из этих архитектур привнесла свои инновации, направленные на улучшение производительности и эффективности. Например, ResNet использовал концепцию остаточных связей, позволяющую тренировать очень глубокие сети без проблем затухания градиента. Эти модели стали стандартом в области компьютерного зрения и были адаптированы для множества приложений.</p>

<p>Одним из важных направлений развития стало применение сверточных нейронных сетей для задач детектирования объектов. Такие модели, как <em>R-CNN</em>, <em>Fast R-CNN</em> и <em>YOLO</em> (You Only Look Once), позволили не только классифицировать объекты на изображениях, но и определять их положение. Это открыло возможности для применения компьютерного зрения в реальном времени, таких как автономное вождение, видеоаналитика и системы безопасности.</p>

<p>Параллельно с этим развивались методы сегментации изображений, позволяющие определять контуры и области объектов с точностью до пикселя. Модели, такие как <em>U-Net</em> и <em>Fully Convolutional Networks</em> (<em>FCN</em>), нашли широкое применение в медицине, сельском хозяйстве и других областях, где важно точно определять границы объектов.</p>

<p>В последние годы наблюдается интерес к моделям генеративного характера, таким как генеративно-состязательные сети (<em>GAN</em>). Эти модели способны создавать новые, реалистичные изображения, обучаясь на больших наборах данных. GAN нашли применение в таких задачах, как улучшение качества изображений, стилизация и восстановление поврежденных данных. Они также открыли новые возможности для изучения и моделирования сложных распределений данных.</p>

<p>Еще одним современным направлением является применение трансформеров в компьютерном зрении. Изначально разработанные для обработки последовательностей в области обработки естественного языка, трансформеры были адаптированы для визуальных данных, что привело к появлению моделей, таких как <em>Vision Transformer</em> (<em>ViT</em>). Эти модели используют механизмы внимания для обработки изображений и показали конкурентные результаты по сравнению с традиционными сверточными сетями.</p>

<p>Развитие методов самообучения и слабого обучения также стало важным трендом. В условиях ограниченности размеченных данных, такие подходы позволяют моделям обучаться на неразмеченных или частично размеченных данных, что расширяет их применимость в реальных условиях. Это особенно важно для областей, где сбор и разметка данных являются дорогостоящими или трудоемкими процессами.</p>

<p>Современные методы компьютерного зрения все чаще интегрируются с другими областями искусственного интеллекта, такими как обработка естественного языка и робототехника. Это приводит к появлению мультидисциплинарных приложений, где системы способны не только видеть, но и понимать контекст, взаимодействовать с окружающей средой и принимать решения. Примером таких систем являются автономные роботы, способные навигировать в сложных условиях, или интеллектуальные ассистенты, анализирующие визуальную информацию и отвечающие на вопросы пользователей.</p>

<p>Важным аспектом развития алгоритмов компьютерного зрения является оптимизация и адаптация моделей для работы на ограниченных ресурсах. С появлением мобильных устройств и внедрением технологий в повседневную жизнь возникла необходимость создавать эффективные и легковесные модели. Методы, такие как сжатие моделей, квантование и прунинг, позволяют уменьшать размер и вычислительную сложность нейронных сетей без существенной потери точности.</p>

<p>Развитие облачных технологий и распределенных вычислений также повлияло на область компьютерного зрения. Возможность обработки больших объемов данных в облаке и использование мощных серверов для обучения моделей ускорило исследования и внедрение новых методов. Однако это также подняло вопросы о приватности данных и безопасности, что стимулировало развитие методов конфиденциального и федеративного обучения.</p>

<p>Эволюция алгоритмов компьютерного зрения тесно связана с развитием аппаратного обеспечения. Появление графических процессоров (<em>GPU</em>) и специальных ускорителей для машинного обучения позволило существенно увеличить скорость обучения и обработки моделей. Это создало условия для разработки и тестирования более сложных и глубоких нейронных сетей, которые ранее были вычислительно неосуществимы.</p>

<p>Необходимо отметить, что развитие компьютерного зрения происходит не только благодаря техническим инновациям, но и благодаря развитию сообществ и открытых источников. <h2>Представление изображений как данных</h2>

<p>Представление изображений как данных является фундаментальным аспектом компьютерного зрения. Понимание того, как изображения хранятся и обрабатываются компьютерами, позволяет эффективно разрабатывать алгоритмы и модели для анализа визуальной информации.

Мы рассмотрим, как изображения представимы в цифровом виде, изучим различные цветовые пространства и обсудим методы предобработки изображений, которые часто используются в задачах компьютерного зрения.</p>

<p>Изображение в цифровом формате представляет собой двумерный массив чисел, где каждое число соответствует пикселю — наименьшему элементу изображения. В случае черно-белых изображений каждый пиксель хранит информацию о яркости в градациях серого, обычно в диапазоне от 0 (черный) до 255 (белый) при 8-битном представлении. Таким образом, черно-белое изображение можно представить как матрицу размером $H \times W$, где $H$ — высота изображения, а $W$ — его ширина.</p>

<p>В цветных изображениях информация о каждом пикселе более сложна. Цветовые изображения обычно представлены в формате RGB, где каждый пиксель характеризуется тремя компонентами: красным (Red), зеленым (Green) и синим (Blue) каналами. Каждая из этих компонент также имеет значение в диапазоне от 0 до 255 при 8-битном представлении. Таким образом, цветное изображение можно представить как трехмерный тензор размером \(H \times W \times 3\), где третий размер соответствует трем цветовым каналам.</p>

<p>Глубина цвета, или битовая глубина, определяет количество битов, используемых для представления цвета каждого пикселя. Чем больше битов используется, тем большее количество оттенков может быть представлено. Например, при 8-битном представлении на канал (общая глубина 24 бита для RGB-изображения) можно отобразить около 16,7 миллионов различных цветов. В некоторых приложениях используется более высокая битовая глубина для обеспечения более точного представления цветов, особенно в профессиональной фотографии и графическом дизайне.</p>

<p>Цветовое пространство RGB является наиболее распространенным и интуитивно понятным, так как основано на принципах аддитивного смешения света. Однако существуют и другие цветовые пространства, которые используются в различных задачах компьютерного зрения и обработки изображений. Одним из таких пространств является HSV (Hue, Saturation, Value), где цвет описывается тоном, насыщенностью и значением (яркостью). Это цветовое пространство более близко соответствует человеческому восприятию цвета и часто используется в задачах, связанных с обработкой цвета, таких как сегментация по цвету или изменение цветовых оттенков.</p>

<p>Еще одним важным цветовым пространством является Lab (Lightness, a*, b*), которое стремится к перцептивной равномерности, то есть одинаковые изменения в значениях компонентов должны приводить к одинаковым изменениям в воспринимаемом цвете. Это полезно в задачах, где необходимо учитывать особенности человеческого восприятия цвета, например, при цветокоррекции или сравнении цветов.</p>

<p>Переход между цветовыми пространствами осуществляется с помощью математических преобразований. Библиотеки для обработки изображений, такие как OpenCV или PIL в Python, предоставляют функции для конвертации изображений из одного цветового пространства в другое. Это позволяет использовать наиболее подходящее цветовое пространство для конкретной задачи.</p>

<p>Предобработка изображений является важным этапом в большинстве приложений компьютерного зрения. Одной из основных целей предобработки является приведение данных к такому виду, который будет наиболее удобен для дальнейшего анализа или обучения моделей. Одним из методов предобработки является нормализация данных, которая заключается в приведении значений пикселей к определенному диапазону, обычно от 0 до 1 или с нулевым средним и единичной дисперсией. Это позволяет избежать проблем, связанных с различиями в масштабах данных, и способствует более стабильному и быстрому обучению моделей машинного обучения.</p>

<p>Стандартизация данных, похожая на нормализацию, включает вычитание среднего значения и деление на стандартное отклонение. Это позволяет получить данные с нулевым средним и единичной дисперсией, что часто улучшает сходимость алгоритмов оптимизации при обучении нейронных сетей.</p>

<p>Масштабирование и изменение размера изображений также являются распространенными операциями предобработки. В задачах, где требуется обработка большого количества изображений, важно, чтобы все изображения имели одинаковый размер и разрешение. Это особенно актуально при обучении нейронных сетей, где размер входных данных должен быть фиксированным. Масштабирование может осуществляться различными методами интерполяции, такими как ближайший сосед, билинейная или бикубическая интерполяция. Каждый метод имеет свои преимущества и недостатки в зависимости от специфики задачи.</p>

<p>Рассмотрим пример использования библиотеки OpenCV для загрузки и предобработки изображения:</p>


```python
# Импорт необходимых библиотек для обработки изображений
import cv2
import numpy as np
from google.colab import files
from matplotlib import pyplot as plt

# Функция для отображения изображения в формате RGB
def show_image(image, title="Image"):
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
# Загрузка изображения с локального компьютера пользователя
# После загрузки изображение сохраняется как 'image.jpg'
uploaded = files.upload()
image_path = next(iter(uploaded.keys()))

# Чтение загруженного изображения в цвете
# Используется функция cv2.imread для загрузки изображения в BGR формате.
# Параметр cv2.IMREAD_COLOR указывает, что изображение будет загружено в цвете.
image = cv2.imread(image_path, cv2.IMREAD_COLOR)

# Отображение загруженного изображения
show_image(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), "Original Image")
```

```python
image.shape
```

```python
# Изменение размера изображения до 224x224 пикселей
# cv2.resize используется для изменения размера изображения до заданных значений (224, 224).
# Интерполяция INTER_LINEAR обеспечивает сглаживание изображения, уменьшая артефакты при уменьшении или увеличении размера.
resized_image = cv2.resize(image, (224, 224), interpolation=cv2.INTER_LINEAR)

# Отображение изображения после изменения размера
show_image(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB), "Resized Image")
```

```python
resized_image.shape
```

```python
# Конвертация изображения из BGR в RGB цветовое пространство
# OpenCV загружает изображения в формате BGR, поэтому используем cv2.cvtColor для преобразования в RGB,
# что является стандартным форматом для большинства библиотек и моделей машинного обучения.
rgb_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)

# Отображение изображения после преобразования в RGB
show_image(rgb_image, "RGB Image")
```

```python
# Нормализация значений пикселей к диапазону [0, 1]
# Значения пикселей делятся на 255.0, чтобы перевести их из диапазона [0, 255] в [0, 1].
# Это улучшает численную стабильность при последующей обработке изображения.
normalized_image = rgb_image / 255.0

# Отображение изображения после нормализации
show_image(normalized_image, "Normalized Image")
```

```python
# Вычисление среднего и стандартного отклонения для каждого канала
# np.mean и np.std рассчитывают статистические характеристики для каждого цветового канала (R, G, B)
# по всем пикселям изображения. Это помогает в стандартизации.
mean = np.mean(normalized_image, axis=(0, 1))
std = np.std(normalized_image, axis=(0, 1))

print("Mean:", mean)
print("Standard Deviation:", std)
```

```python
# Стандартизация изображения
# Вычитание среднего и деление на стандартное отклонение для каждого канала позволяет стандартизировать
# изображение. Это делает данные более однородными, что способствует лучшему обучению модели, устраняя
# различия в освещении и цветовых оттенках.
standardized_image = (normalized_image - mean) / std

# Отображение изображения после стандартизации
show_image(standardized_image, "Standardized Image")
```

<p>В этом примере мы сначала загружаем изображение и изменяем его размер до $224 \times 224$ пикселей, что является стандартным размером для многих моделей нейронных сетей, таких как VGG или ResNet. Затем мы конвертируем изображение из цветового пространства BGR, которое используется в OpenCV по умолчанию, в RGB, более распространенное в других библиотеках. После этого нормализуем значения пикселей к диапазону от $0$ до $1$ и проводим стандартизацию по каждому каналу.</p>

<p>Предобработка изображений может также включать другие операции, такие как удаление шума с помощью фильтров, выравнивание гистограммы для улучшения контрастности или применение аугментации данных. Аугментация данных включает в себя искусственное увеличение набора данных путем применения различных преобразований к изображениям, таких как повороты, сдвиги, зеркальные отражения или изменение яркости. Это помогает моделям стать более устойчивыми к вариациям в данных и улучшает их способность к обобщению.</p>

<p>Цветовые пространства играют важную роль в задачах компьютерного зрения. В некоторых случаях переход в другое цветовое пространство может значительно упростить задачу. Например, при детектировании объектов определенного цвета может быть полезно использовать цветовое пространство HSV, где оттенок цвета отделен от его яркости и насыщенности. Это позволяет более эффективно выделять нужные объекты независимо от условий освещения.</p>

<p>Рассмотрим пример конвертации изображения в цветовое пространство HSV и выделения пикселей определенного оттенка:</p>

```python
# Загрузка изображения с локального компьютера пользователя
# После загрузки изображение сохраняется как 'image.jpg'
uploaded = files.upload()
image_path = next(iter(uploaded.keys()))

# Чтение загруженного изображения в цвете
# Используется функция cv2.imread для загрузки изображения в BGR формате.
image = cv2.imread(image_path, cv2.IMREAD_COLOR)

# Отображение загруженного изображения
show_image(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), "Original Image")
```

```python
# Конвертация изображения из BGR в HSV цветовое пространство
# OpenCV поддерживает преобразование цветовых пространств, и в данном случае мы конвертируем изображение в HSV,
# что облегчает работу с отдельными цветами.
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Отображение изображения в формате HSV
show_image(cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB), "HSV Image")
```

```python
# Анализируем диапазон значений в HSV-каналах изображения для точной настройки маски
# Получаем минимальные и максимальные значения для каждого канала
h_channel_min, h_channel_max = hsv_image[:, :, 0].min(), hsv_image[:, :, 0].max()
s_channel_min, s_channel_max = hsv_image[:, :, 1].min(), hsv_image[:, :, 1].max()
v_channel_min, v_channel_max = hsv_image[:, :, 2].min(), hsv_image[:, :, 2].max()

(h_channel_min, h_channel_max), (s_channel_min, s_channel_max), (v_channel_min, v_channel_max)
```

```python
# Пример диапазона для выделения теплых оттенков (например, рыжего или коричневого)
# В диапазоне задаются значения оттенка, насыщенности и яркости, чтобы выделить  цвет.
lower_gray = np.array([0, 0, 150])
upper_gray = np.array([150, 20, 255])

# Создание маски на основе заданного диапазона
mask = cv2.inRange(hsv_image, lower_gray, upper_gray)

# Отображение маски
show_image(mask, "Mask")
```

```python
# Применение маски для выделения цвета на исходном изображении
# Функция bitwise_and оставляет только те пиксели, которые находятся в диапазоне синего цвета.
result = cv2.bitwise_and(image, image, mask=mask)

# Отображение результата с выделением  цвета
show_image(cv2.cvtColor(result, cv2.COLOR_BGR2RGB), "Image with Color Highlighted")
```

<p>В этом примере мы конвертируем изображение в цветовое пространство HSV и создаем маску, которая выделяет пиксели с оттенками указанного цвета в заданном диапазоне. Затем мы применяем эту маску к исходному изображению, получая изображение, где отображаются только нужные нам области.</p>

<p>Понимание различных цветовых пространств и их особенностей позволяет выбирать наиболее подходящие методы для решения конкретных задач. Например, цветовое пространство Lab часто используется в задачах цветовой сегментации и коррекции, так как оно более точно отражает человеческое восприятие цвета.</p>

<p>Важным аспектом представления изображений является их формат хранения и сжатия. Изображения могут быть сохранены в различных форматах, таких как JPEG, PNG, BMP и другие. Каждый формат имеет свои особенности, связанные с качеством изображения, размером файла и возможностью сохранения прозрачности или анимации. При обработке изображений важно учитывать, что некоторые форматы используют сжатие с потерями, что может повлиять на качество и точность последующей обработки.</p>

<p>При работе с большими наборами данных изображений важно также учитывать организацию данных и эффективность их загрузки и обработки. Использование эффективных форматов хранения, таких как TFRecords в TensorFlow или HDF5, может значительно ускорить процесс обучения моделей.</p>

<p>Еще одним важным аспектом является управление каналами изображения. В некоторых задачах может потребоваться работа не только с тремя цветовыми каналами, но и с дополнительными информационными каналами, такими как альфа-канал (прозрачность) или каналы глубины и инфракрасного излучения в специализированных приложениях. Умение работать с многоканальными изображениями расширяет возможности анализа и обработки данных.</p>

<p>В задачах машинного обучения на основе изображений, особенно при использовании глубоких нейронных сетей, важным этапом является подготовка данных. Помимо нормализации и стандартизации, часто используется централизованная нормализация, где из каждого канала вычитается среднее значение по обучающему набору данных, и делится на стандартное отклонение. Это позволяет учесть статистические свойства всего набора данных и улучшить качество обучения.</p>

<p>Например, при обучении моделей на наборе данных ImageNet часто используются следующие значения для нормализации:</p>

<pre>
<code>
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]

# Нормализация изображения
normalized_image = (image - mean) / std
</code>
</pre>

<p>Здесь среднее и стандартное отклонение вычислены по всему набору данных ImageNet и используются для приведения новых изображений к тому же статистическому распределению.</p>

<p>Важно также учитывать порядок каналов в изображении. Некоторые библиотеки, такие как TensorFlow, используют формат данных с каналами последними (NHWC), где данные представлены как тензор размером $(N, H, W, C)$, где $N$ — размер батча, $H$ — высота, $W$ — ширина, $C$ — число каналов. Другие библиотеки, такие как PyTorch, используют формат с каналами первыми (NCHW). При переносе моделей или данных между разными фреймворками необходимо учитывать эти различия и при необходимости переставлять оси тензоров.</p>

<p>Кроме того, при работе с изображениями важно учитывать вопрос аугментации данных. Аугментация позволяет увеличить размер набора данных и улучшить обобщающую способность модели. Примеры аугментации включают случайные повороты, сдвиги, изменение яркости и контраста, добавление шума и другие преобразования. В современных библиотеках, таких как Keras или PyTorch, существуют готовые функции и классы для реализации аугментации данных.</p>

<p>Рассмотрим пример применения аугментации данных с использованием библиотеки torchvision в PyTorch:</p>

```python
# Импорт необходимых библиотек для работы с изображениями и преобразований
import cv2
import numpy as np
import torch
from torchvision import transforms
from google.colab import files
from PIL import Image
from matplotlib import pyplot as plt

# Функция для отображения изображения в формате RGB
def show_image(image, title="Image"):
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
# Загрузка изображения с локального компьютера пользователя
# После загрузки изображение сохраняется как 'image.jpg'
uploaded = files.upload()
image_path = next(iter(uploaded.keys()))

# Чтение загруженного изображения в цвете
# Используется функция cv2.imread для загрузки изображения в BGR формате.
# Параметр cv2.IMREAD_COLOR указывает, что изображение будет загружено в цвете.
image = cv2.imread(image_path, cv2.IMREAD_COLOR)

# Преобразование изображения из BGR в RGB и его отображение
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
show_image(rgb_image, "Original Image")

# Преобразование изображения в формат PIL для последующих шагов
pil_image = Image.fromarray(rgb_image)
```

```python
# Вычисление среднего и стандартного отклонения для каждого канала
# Эти значения будут использованы для нормализации изображения в преобразованиях.
mean = np.mean(rgb_image / 255.0, axis=(0, 1))
std = np.std(rgb_image / 255.0, axis=(0, 1))

print("Mean:", mean)
print("Standard Deviation:", std)
```

```python
# Определение последовательности преобразований, которые будут применены к изображению
# transforms.Compose объединяет несколько преобразований, которые выполняются последовательно.
data_transforms = transforms.Compose([
    transforms.RandomResizedCrop(224), # Рандомное изменение размера и обрезка до 224x224 пикселей
    transforms.RandomHorizontalFlip(), # Случайное горизонтальное отражение
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1), # Рандомные изменения цвета
    transforms.ToTensor(), # Преобразование изображения в тензор
    transforms.Normalize(mean, std) # Нормализация с использованием вычисленных среднего и стандартного отклонения
])

# Применение преобразований к изображению
augmented_image = data_transforms(pil_image)
```

```python
# Поскольку изображение было преобразовано в тензор, его необходимо преобразовать обратно для отображения
# Сначала денормализуем изображение, затем преобразуем из тензора в формат numpy для визуализации
def denormalize(tensor_image, mean, std):
    # Восстановление нормализованных значений изображения, умножив на стандартное отклонение и прибавив среднее
    for t, m, s in zip(tensor_image, mean, std):
        t.mul_(s).add_(m)
    # Обрезаем значения, чтобы они находились в диапазоне [0, 1]
    return torch.clamp(tensor_image, 0, 1)

# Денормализуем изображение для визуализации
denormalized_image = denormalize(augmented_image.clone(), mean, std)

# Преобразуем тензор обратно в numpy для отображения
denormalized_image_np = denormalized_image.permute(1, 2, 0).numpy()
show_image(denormalized_image_np, "Augmented Image")
```

<p>Здесь мы определяем последовательность преобразований, которые включают случайное обрезание и изменение размера изображения, случайное горизонтальное отражение, случайное изменение яркости, контраста, насыщенности и оттенка, преобразование изображения в тензор и нормализацию. Это позволяет генерировать различные варианты изображений на основе исходного, что улучшает устойчивость модели к разнообразию входных данных.</p>

<p>Представление изображений как данных — это не только вопрос технического хранения и обработки, но и понимание того, как информация, содержащаяся в изображениях, может быть эффективно использована в моделях машинного обучения. Глубокие нейронные сети, особенно сверточные нейронные сети (CNN), были разработаны с учетом пространственной структуры изображений. Они используют локальные рецептивные поля и разделение параметров для эффективного обучения признаков на изображениях.</p>

<p>Понимание того, как изображения представлены и предобрабатываются, является ключевым для успешного применения методов компьютерного зрения. Это позволяет выбирать правильные подходы к обработке данных, настраивать модели и добиваться лучших результатов в задачах классификации, детектирования, сегментации и других областях.</p>

<h2>Извлечение признаков</h2>


<p>Извлечение признаков из изображений является еще одной фундаментальной задачей в компьютерном зрении, позволяющей преобразовывать визуальные данные в численные представления, пригодные для последующей обработки и анализа. Это критически важно для таких приложений, как распознавание объектов, отслеживание движений и 3D-реконструкция. Мы рассмотрим некоторые методы извлечения признаков, такие как гистограммы направленных градиентов (HOG) и SIFT (Scale-Invariant Feature Transform), а также обсудим понятие пространственно-инвариантных признаков и сравним различные подходы.</p>

<h3>Гистограммы градиентов (HOG)</h3>

<p>Гистограмма направленных градиентов (HOG) — это метод описания локальных объектов на изображении, основанный на распределении градиентов или краев. Основная идея HOG заключается в том, что локальная форма объекта может быть хорошо охарактеризована распределением направлений градиентов интенсивности, даже без точной информации о положении краев.</p>

<h4>Что такое дескриптор признаков?</h4>

<p>Дескриптор признаков — это представление изображения или фрагмента изображения, которое упрощает изображение, извлекая полезную и отбрасывая бесполезную информацию.</p>

<p>HOG-дескриптор преобразует изображение размером
ширина $H \times W \times 3$ (три канала цвета) в одномерный вектор признаков длиной $n$. Например, входное изображение размером $64 \times 128 \times 3$ преобразуется в вектор признаков длиной $3780$ элементов. Этот вектор содержит упрощенную информацию об изображении, фокусируясь на важных деталях, таких как края и контуры, и игнорируя менее значимые детали, например, цветовые вариации.</p>

<p><strong>HOG может быть рассчитан для других размеров.</strong></p>

<h4>Какие признаки полезны?</h4>

<p>В задачах распознавания изображений и обнаружения объектов важными считаются признаки, которые позволяют отличать один объект от другого. Например, для детектирования пуговиц на одежде важны круглые формы и наличие отверстий, а цвет может быть менее значимым. В дескрипторе признаков HOG в качестве признаков используются распределения (гистограммы) направлений градиентов (ориентированные градиенты). Градиенты (производные по $x$  и  $y$) изображения полезны, потому что величина градиентов велика по краям и углам (области резкого изменения интенсивности), и мы знаем, что края и углы содержат гораздо больше информации о форме объекта, чем плоские области.</p>

<h4>Как рассчитать гистограмму ориентированных градиентов?</h4>

<p>1. Предварительная обработка.<br>
Как упоминалось ранее, дескриптор функции HOG, используемый для обнаружения пешеходов, рассчитывается на участке изображения размером $64 \times 128$ пикселей. Конечно, изображение может быть любого размера. Обычно пятна в нескольких масштабах анализируются во многих местах изображения. Единственное ограничение заключается в том, что анализируемые фрагменты имеют фиксированное соотношение сторон. В нашем случае, фрагменты должны иметь соотношение сторон $1:2$. Например, они могут быть $100 \times 200$,$128 \times 256$ или $1000 \times 2000$, но не $101 \times 205$.</p>

<p>Чтобы проиллюстрировать этот момент, рассмотрим большое изображение размером $729 \times 475$. Был выбран фрагмент размером $100 \times 200$ для вычисления нашего дескриптора признака HOG. Этот фрагмент вырезается из изображения и изменяется до $64 \times 128$. Теперь мы можем вычислить дескриптор HOG для этого фрагмента изображения.</p>

<center> <img src="https://drive.google.com/uc?id=13LhCLRJrSL7R-91NVBJIfCknEqoYJI9z" alt="" width="800"/>
<p> Рисунок 2 — Обрезка изображения </p> </center>

<p>2. Градиент изображения.<br>
Чтобы вычислить дескриптор HOG, нужно вычислить горизонтальный и вертикальный градиенты. Это достигается путем фильтрации изображения с помощью следующих ядер.</p>

$$\text{Ядро для } G_x \text{: }\begin {bmatrix} -1 & 0 & 1 \end{bmatrix},$$

$$\text{Ядро для } G_y \text{: }\begin {bmatrix} -1 \\ 0 \\ 1 \end{bmatrix}.$$

<p>В OpenCV это можно реализовать с помощью оператора Собеля с размером ядра $1 \times 1$:</p>

<pre>
<code>
# Чтение изображения и преобразование в оттенки серого
img = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)
img = np.float32(img) / 255.0

# Вычисление градиентов по x и y
gx = cv2.Sobel(img, cv2.CV_32F, 1, 0, ksize=1)
gy = cv2.Sobel(img, cv2.CV_32F, 0, 1, ksize=1)
</code>
</pre>

<p>Формально, градиент изображения в каждой точке определяет направление наибольшего увеличения интенсивности и вычисляется с использованием производных по $x$ и $y$</p>

$$G_x = I(x+1, y) - I(x-1, y),$$
$$G_y = I(x, y+1) - I(x, y-1),$$

<p>где $G_x$ и $G_y$ — горизонтальные и вертикальные компоненты градиента;<br>
$I(x, y)$ — интенсивность пикселя в точке $(x, y)$.</p>

<p>Модуль и направление градиента вычисляются как</p>

$$|G| = \sqrt{G_x^2 + G_y^2},$$
$$\theta = \arctan{\frac{G_x}{G_y}}.$$

<p>Используя OpenCV, расчет можно выполнить с помощью функции cartToPolar:</p>

<pre>
<code>
# Вычислить величину и направление градиента (в градусах)
mag, angle = cv2.cartToPolar(gx, gy, angleInDegrees=True)
</code>
</pre>

<center> <img src="https://drive.google.com/uc?id=17lQmnyCfffn-jpG9owmpvGM_sZTQnNmt" alt="" width="800"/>
<p> Рисунок 3 — Градиенты: слева - абсолютное значение градиента $G_x$, в центре - абсолютное значение градиента $G_y$, справа - величина градиента $|G|$.</p> </center> </p>

<p>Обратите внимание, что градиент $G_x$ проявляется на вертикальных линиях, а градиент $G_y$ виден на горизонтальных линиях. Вспышки величины градиентов везде, где наблюдается резкое изменение интенсивности. Ни градиент $G_x$, ни градиент $G_y$ не наблюдаются на гладких областях.</p>

<p>Градиентное изображение удалило много несущественной информации (например, постоянный цветной фон), но выделило контуры. Другими словами, вы можете смотреть на градиентное изображение и все равно легко сказать, что на нем изображен человек.</p>

<p>В каждом пикселе градиент имеет величину и направление. Для цветных изображений оцениваются градиенты трех каналов (как показано на рисунке выше). Величина градиента в пикселе — это максимум величины градиентов трех каналов, а угол — это угол, соответствующий максимальному градиенту.</p>

<p>3. Вычисление гистограммы.<br>
На этом этапе изображение делится на ячейки $8 \times 8$ и гистограмма градиентов вычисляется для каждой ячейки $8 \times 8$.
</p>

<p>Одна из важных причин использования дескриптора функции для описания фрагмента изображения заключается в том, что он обеспечивает компактное представление. Патч изображения $8 \times 8$ содержит значения $8 \cdot 8 \cdot 3 = 192$ пикселей. Градиент этого патча содержит $2$ значения (величину и направление) на пиксель, что в сумме дает $8 \cdot 8 \cdot 2 = 128$ чисел. Позже мы увидим, как эти $128$ чисел представлены с использованием гистограммы с $9$ ячейками, которую можно сохранить как массив из $9$ чисел. Мало того, что представление становится более компактным,так и вычисление гистограммы по фрагменту делает это представление более устойчивым к шуму. На отдельных участках может быть шум, но гистограмма на участке $8 \times 8$ делает изображение менее чувствительным к шуму.</p>

<p>Но почему патч $8 \times 8$? Размер ячеек выбран эмпирически и соответствует масштабу интересующих признаков (например, частей тела пешехода). Это делается для упрощения обработки и создания локальных гистограмм градиентов.</p>

<p>Справа мы видим необработанные числа, представляющие градиенты в ячейках $8 \times 8$ с одним незначительным отличием — углы составляют от $0$ до $180$ градусов вместо $0$ до $360$ градусов. Они называются «беззнаковыми» градиентами, потому что градиент и отрицательный градиент представлены одними и теми же числами. Другими словами, градиентная стрелка и стрелка на $180$ градусов, противоположная ей, считаются одинаковыми. Но почему бы не использовать угол $0$ — $360$ градусов? Но некоторые реализации HOG позволят вам указать, хотите ли вы использовать подписанные градиенты.</p>

Так как $10$ градусов находятся на полпути между $0$ и $20$, голос пикселя равномерно распределяется на две ячейки</p>

<p>4. Нормаолизация блока $16 \times 16$.
<br>
На предыдущем шаге мы создали гистограмму на основе градиента изображения. Градиенты изображения чувствительны к общему освещению. Если сделать изображение темнее, разделив все значения пикселей на $2$, величина градиента изменится вдвое, и, следовательно, значения гистограммы изменятся вдвое.
<br>
В идеале, мы хотим, чтобы наш дескриптор не зависел от вариаций освещения. Другими словами, мы хотели бы «нормализовать» гистограмму, чтобы на них не влияли изменения освещения.
<br>
Рассмотрим, как нормализуется вектор длины 3 в качестве примера.
<br>
Допустим, у нас есть цветовой вектор RGB $[128, 64, 32]$. Длина этого вектора равна $\sqrt{128^2+64^2+32^2} = 146,64$. Это называется L2-нормой вектора. Разделив каждый элемент этого вектора на $146,64$, мы получим нормализованный вектор $[0,87, 0,43, 0,22]$.
<br>
Теперь рассмотрим другой вектор, в котором элементы в два раза превышают значение первого вектора $2 \cdot [128, 64, 32] = [256, 128, 64]$. В таком случае нормализация вектора $[256, 128, 64]$ приведет к $[0,87, 0,43, 0,22]$, что совпадает с нормализованной версией исходного вектора RGB. Вы можете видеть, что нормализация вектора удаляет масштаб.
<br>
Теперь, когда мы знаем, как нормализовать вектор, мы можем подумать, что при вычислении HOG возможно просто нормализовать гистограмму $9 \times 1$ так же, как мы нормализовали вектор $3 \times 1$ выше. Это неплохая идея, но лучше нормализовать по блоку большего размера $16 \times 16$. Блок $16 \times 16$ имеет $4$ гистограммы, что могут быть объединены, чтобы сформировать вектор элементов $36 \times 1$, и его можно нормализовать так же, как нормализован вектор $3 \times 1$. Затем окно перемещается на $8$ пикселей (рисунок 8), и для этого окна вычисляется нормализованный вектор размером $36 \times 1$, и процесс повторяется.
</p>

<p>5. Гистограмма векторов ориентированного градиента.
<br>
Чтобы вычислить окончательный вектор признаков для всего фрагмента изображения, векторы размером $36 \times 1$ объединяются в один гигантский вектор.
<br>
Есть $7$ горизонтальных и $15$ вертикальных позиций, что в сумме составляет $7 \cdot 15 = 105$ позиций.
<br>
Каждый блок $16 \times 16$ представлен вектором $36 \times 1$. Итак, когда мы объединяем их все в один вектор затухания, мы получаем размерный вектор $36 \times 105 = 3780$.
</p>

<p>После того как мы получили итоговый вектор признаков, следующий шаг — использовать этот вектор для обучения модели машинного обучения, обычно классификатора, такого как метод опорных векторов (SVM). Цель состоит в том, чтобы научить модель распознавать определенные объекты или классифицировать изображения на основе извлеченных HOG-признаков.</p>

<p>Для начала мы можем взять стандартное изображение из библиотеки scikit-image и извлечь из него HOG-признаки. Затем визуализируем эти признаки и обсудим, как они могут быть использованы в дальнейшем.</p>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, exposure
from skimage.feature import hog

# Загрузка стандартного изображения из scikit-image
# Используется встроенное изображение "астронавт" из scikit-image, которое предоставляет цветное изображение.
image = data.astronaut()
image = cv2.resize(image, (128, 64))  # Изменяем размер изображения для соответствия требованиям HOG

# Преобразование в оттенки серого
# HOG лучше работает с изображениями в оттенках серого, поэтому преобразуем изображение с помощью cv2.cvtColor.
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Вычисление HOG-признаков и получение изображения с визуализацией
# Настройка параметров для извлечения HOG (Histogram of Oriented Gradients) признаков:
# - orientations=9: количество направлений градиента (9 ориентированных направлений)
# - pixels_per_cell=(8, 8): размер ячейки в пикселях, где вычисляются градиенты
# - cells_per_block=(2, 2): блок из 2х2 ячеек, где нормализуются гистограммы
# - block_norm='L2-Hys': нормализация для стабильности признаков
# - visualize=True: возвращает изображение, показывающее визуализацию HOG признаков
# - transform_sqrt=True: улучшает контрастность изображения перед HOG
# - feature_vector=True: возвращает признаки как одномерный массив, готовый для использования в модели
hog_features, hog_image = hog(
    gray_image,
    orientations=9,
    pixels_per_cell=(8, 8),
    cells_per_block=(2, 2),
    block_norm='L2-Hys',
    visualize=True,
    transform_sqrt=True,
    feature_vector=True,
)

# Вывод размеров вектора признаков
# Печатает форму HOG-признаков, которая показывает общее количество извлеченных признаков,
# что важно для анализа размерности данных в последующих моделях.
print(f'Размер вектора признаков: {hog_features.shape}')

# Визуализация исходного изображения и HOG-признаков
# Создаем два подграфика для отображения исходного изображения и визуализации HOG-признаков
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6), sharex=True, sharey=True)

# Отображение исходного изображения
ax1.axis('off')  # Отключаем оси для улучшения визуального восприятия
ax1.imshow(gray_image, cmap=plt.cm.gray)  # Отображаем изображение в оттенках серого
ax1.set_title('Исходное изображение')

# Улучшаем контрастность для визуализации
# Используем rescale_intensity из skimage.exposure для масштабирования интенсивности HOG-изображения,
# чтобы лучше отобразить контрасты градиентов.
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))

# Отображение HOG-признаков
ax2.axis('off')  # Отключаем оси
ax2.imshow(hog_image_rescaled, cmap=plt.cm.gray)  # Отображаем визуализацию HOG-признаков
ax2.set_title('HOG-признаки')

# Выводим оба изображения
plt.show()
```

<p>Вектор признаков HOG размером $3780$ элементов является представлением изображения в пространстве признаков. Каждый элемент этого вектора соответствует определенной гистограмме ориентированных градиентов в определенном блоке изображения.</p>

<p>Даже без обучения классификатора вы можете анализировать и сравнивать эти векторы признаков между разными изображениями, чтобы понять, как различия в изображениях отражаются в пространстве признаков.</p>

<p>Далее возможно взять два разных изображения и вычислить для них HOG-признаки. Затем можно вычислить расстояние между их векторами признаков (например, с помощью евклидова расстояния или косинусного сходства) для оценки степени их похожести. Это может быть полезно для задач, связанных с поиском похожих изображений или кластеризацией без необходимости обучения модели.</p>

```python
# Предположим, у нас есть второе изображение
# Используем встроенное изображение "camera" из scikit-image в качестве второго изображения для анализа
image2 = data.camera()
image2 = cv2.resize(image2, (128, 64))  # Изменяем размер изображения для соответствия требованиям HOG

# Вычисляем HOG-признаки для второго изображения
# Повторяем процесс HOG, используя те же параметры для консистентности признаков между изображениями.
# Это позволяет впоследствии корректно сравнивать их по извлеченным признакам.
hog_features2, _ = hog(
    image2,
    orientations=9,
    pixels_per_cell=(8, 8),
    cells_per_block=(2, 2),
    block_norm='L2-Hys',
    visualize=True,
    transform_sqrt=True,
    feature_vector=True,
)

# Вычисляем косинусное сходство между двумя векторами признаков
# Используем косинусное сходство (cosine similarity) из sklearn для измерения степени сходства между
# векторами признаков hog_features и hog_features2. Косинусное сходство является распространенной мерой,
# особенно подходящей для высокоразмерных признаковых пространств, таких как HOG.
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(
    hog_features.reshape(1, -1),   # Преобразуем первый вектор признаков в двумерный массив
    hog_features2.reshape(1, -1)   # Преобразуем второй вектор признаков в двумерный массив
)

# Визуализация второго изображения
# Отображаем второе изображение в оттенках серого для наглядности
plt.axis('off')
plt.imshow(image2, cmap='gray')
plt.show()

# Выводим косинусное сходство между изображениями
# Печатаем значение косинусного сходства, которое варьируется от -1 (совершенно разные) до 1 (идентичные).
# Чем ближе значение к 1, тем больше схожести между HOG-признаками изображений.
print(f'Косинусное сходство между изображениями: {similarity[0][0]}')
```

<p>Если у нас есть несколько изображений, мы можем вычислить для них HOG-признаки и затем использовать методы кластеризации (например, K-Means) для группировки похожих изображений. Это полезно для организации данных и выявления скрытых структур в наборе изображений.</p>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, exposure
from skimage.feature import hog

# Список изображений из skimage.data
# Здесь загружаем четыре изображения из библиотеки scikit-image для демонстрации работы с различными изображениями.
images = [data.astronaut(), data.camera(), data.coins(), data.horse()]
hog_features_list = []  # Инициализируем список для хранения HOG-признаков каждого изображения

for img in images:
    # Проверяем тип данных изображения
    # Некоторые изображения могут быть в булевом формате (например, бинарные маски), который необходимо
    # преобразовать в тип uint8 для корректной обработки.
    if img.dtype == 'bool':
        # Преобразуем булевое изображение в uint8, умножив на 255, чтобы получить стандартные значения интенсивности.
        img = img.astype('uint8') * 255
    else:
        # Убеждаемся, что изображение уже имеет тип uint8 для совместимости с OpenCV
        img = img.astype('uint8')

    # Изменяем размер изображения
    # Преобразуем изображение к размеру 128x64 пикселей, подходящему для HOG, для унификации обработки
    img_resized = cv2.resize(img, (128, 64))

    # Преобразуем в оттенки серого, если необходимо
    # HOG алгоритм требует изображения в оттенках серого, поэтому если изображение цветное (3 канала),
    # преобразуем его, иначе оставляем как есть.
    if len(img_resized.shape) == 3:
        img_gray = cv2.cvtColor(img_resized, cv2.COLOR_RGB2GRAY)
    else:
        img_gray = img_resized

    # Вычисляем HOG-признаки
    # Используем те же параметры HOG, чтобы обеспечить консистентное извлечение признаков:
    # - orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2) для детальной структуры,
    # - block_norm='L2-Hys' для нормализации, transform_sqrt=True для контрастности.
    features, hog_image = hog(
        img_gray,
        orientations=9,
        pixels_per_cell=(8, 8),
        cells_per_block=(2, 2),
        block_norm='L2-Hys',
        visualize=True,
        transform_sqrt=True,
        feature_vector=True,
    )
    hog_features_list.append(features)  # Сохраняем вектор признаков для анализа или дальнейшего использования

    # Визуализация HOG-признаков (опционально)
    # Масштабируем интенсивность HOG-изображения для визуализации, чтобы увидеть детали признаков.
    hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))
    plt.figure(figsize=(8, 4))

    # Отображаем исходное изображение в оттенках серого
    plt.subplot(1, 2, 1)
    plt.imshow(img_gray, cmap='gray')
    plt.title('Изображение в оттенках серого')
    plt.axis('off')

    # Отображаем визуализацию HOG-признаков
    plt.subplot(1, 2, 2)
    plt.imshow(hog_image_rescaled, cmap='gray')
    plt.title('Визуализация HOG')
    plt.axis('off')

    # Показываем изображения на экране
    plt.show()
```
