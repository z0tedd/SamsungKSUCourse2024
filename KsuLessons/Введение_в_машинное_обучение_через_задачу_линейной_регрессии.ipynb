<h1>1. Основные понятия машинного обучения</h1>

<h2>Что такое машинное обучение?</h2>

<p>Машинное обучение — это направление искусственного интеллекта, которое фокусируется на разработке алгоритмов, способных учиться наданных, делать прогнозы и принимать решения,не будучи явно запрограммированными для каждой конкретной задачи.</p>

<p>Основные идеи<ul>
  <li>Обучение на данных — алгоритм анализирует  обучающие данные и выявляет в них закономерности</li>
  <li>Обобщение — алгоритм применяет выявленные закономерности к новым данным, с которыми ранее не работал</li>
</ul></p>

<h2>Классификация способов машинного обучения</h2>

<p>Обучение с учителем (Supervised Learning) —  модель обучается на размеченных данных, где каждой входной последовательности соответствует известный правильный выход, а целью является - научиться предсказывать значения для новых входных данных. Это может включать предсказание непрерывной величины (задача регрессии) или же присвоение входным данным одного из нескольких дискретных классов (задача классификации).</p>

<p>Обучения без учителя (Unsupervised Learning) — модель работает с неразмеченными данными и пытается выявить внутренние закономерности, чтобы  найти скрытые структуры или распределения в данных. Сюда относятся такие задачи, как группировки похожих объектов (кластеризации) или выявления закономерностей совместного появления элементов (ассоциативные правила).</p>

<p>Обучение с подкреплением (Reinforcement learning) — модель обучается, взаимодействуя с окружающей средой, чтобы разработать стратегии, максимизирующие общее вознаграждение, получая награды или наказания за свои действия. Сюда относятся такие задачи, как обучение роботов пространственной навигации или обучение агентов играм.</p>

<h2>Основные компоненты системы машинного обучения</h2>

<p><ul>
  <li>Признаки — свойства или атрибуты объекта, используемые для обучения модели</li>
  <li>Целевая переменная — значение, что модель должна предсказать</li>
  <li>Модель — математическая или алгоритмическая структура, описывающая зависимость между признаками и целевой переменной</li>
  <li>Параметры модели — настраиваемые величины, определяющие поведение модели</li>
  <li>Функция потерь (Loss Function) — функция, определяющая меру несоответствия между предсказанными моделью значениями и реальными</li>
</ul></p>

<h2>Постановка задачи</h2>

<p>Цель: разработать модель для оценки рыночной стоимости на основе данных характеристик жилья.</p>

<p>Предположим, имеется набор данных с <em>m</em> объектов (домов), где каждый объект определяется набором свойств:</p>

<div style="text-align:center;">
  $$ x^{(i)} = [x_1^{(i)}, x_2^{(i)}, \dots, x_n^{(i)}] $$
</div>

<p>где \( x_1^{(i)}, x_2^{(i)}, \dots, x_n^{(i)} \) — это различные характеристики дома, а также известное значение цены дома \( y^{(i)} \).</p>

<h3>Задача регрессии</h3>

<blockquote><p>Отображе́ние в математике — это закон, по которому каждому элементу \( x \) некоторого заданного множества \( X \) сопоставляется определённый элемент \( y \) другого заданного множества \( Y \) (при этом \( X \) может совпадать с \( Y \)). Такое соотношение между элементами \( x \in X \) и \( y \in Y \) записывается в виде</p>
  
  <div style="text-align:center;">
    $$ y = f(x), \quad y = fx, \quad y = x f \quad \text{или} \quad f : x \mapsto y. $$
  </div>

  <p>Часто говорят, что отображение \( f \) действует из \( X \) в \( Y \) и пишут \( f : X \to Y \) или \( X \xrightarrow{f} Y \). Вместо термина «отображение» часто употребляют термин «оператор» (особенно в функциональном анализе и линейной алгебре), а также «функция» (особенно в случае, когда \( Y \) — числовое множество).</p>

  <p>Отображение \( f : X \to X \) называют также преобразованием множества \( X \).</p>
</blockquote>

<p>Требуется найти функцию</p>

<div style="text-align:center;">
  $$ f: \mathbb{R}^n \to \mathbb{R}. $$
</div>

<p>Функция будет принимать на вход объект \( x \) и предсказывать его рыночную стоимость \( y \)</p>

<div style="text-align:center;">
  $$ \hat{y} = f(x). $$
</div>

<h3>Почему линейная регрессия?</h3>
<p>Линейная регрессия предполагает линейную зависимость между признаками объекта и целевой переменной. Она описывается уравнением</p>

<div style="text-align:center;">
  $$ \hat{y} = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \dots + \theta_n x_n, $$
</div>

<p>где
<ul>
  <li> \( \hat{y} \) — предсказанное значение цены дома,</li>
  <li> \( \theta_0 \) — свободный член (смещение),</li>
  <li> \( \theta_1, \theta_2, \dots, \theta_n \) — коэффициенты модели, описывающие влияние каждого признака.</li>
</ul></p>

<h3>Преимущества линейной регрессии</h3>

<p><ul>
  <li>Легко понять влияние каждого атрибута на конечную цену</li>
  <li>Вычислительная эффективность для больших наборов данных</li>
  <li>Используется в качестве отправной точки для более сложных моделей</li>
</ul></p>

<h3>Связь с машинным обучением</h3>

<p><ul>
  <li>Машинное обучение — область искусственного интеллекта, в которой алгоритмы обучаются на основе данных, чтобы улучшить свои результаты</li>
  <li>Задача регрессии является одной из фундаментальных задач машинного обучения и направлена на прогнозирование непрерывной целевой переменной</li>
  <li>Линейная регрессия — важный базовый метод регрессии для понимания более сложных алгоритмов</li>
</ul></p>

<h1>3. Линейная регрессия: теоретические основы</h1>

<h2>Математическая модель линейной регрессии</h2>

<p>Линейная регрессия предполагает наличие зависимости между входными признаками \( x = [x_1, x_2, \dots, x_n]^\top \) и целевой переменной \( {y} \), которая может быть определена следующим образом </p>

<div style="text-align:center;">
  $$ \hat{y} = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \dots + \theta_n x_n. $$
</div>

<p>Для удобства вычислений модель может быть записана в векторно-матричном представлении</p>

<div style="text-align:center;">
  $$ \hat{y} = X \theta, $$
</div>

<p>где</p>
<ul>
  <li>\( X \) — матрица признаков размерности \( m \times (n+1) \), где \( m \) — число наблюдений, а первый столбец состоит из единиц (для свободного члена \( \theta_0 \)): </li>
</ul>

<div style="text-align:center;">
  $$ X = \begin{bmatrix}
  1 & x_1^{(1)} & x_2^{(1)} & \dots & x_n^{(1)} \\
  1 & x_1^{(2)} & x_2^{(2)} & \dots & x_n^{(2)} \\
  \vdots & \vdots & \vdots & \ddots & \vdots \\
  1 & x_1^{(m)} & x_2^{(m)} & \dots & x_n^{(m)}
  \end{bmatrix} $$
</div>

<ul>
  <li>\( \theta \) — вектор параметров размерности \( (n+1) \times 1 \):</li>
</ul>

<div style="text-align:center;">
  $$ \theta = \begin{bmatrix}
  \theta_0 \\
  \theta_1 \\
  \vdots \\
  \theta_n
  \end{bmatrix} $$
</div>

<ul>
  <li>\( \hat{y} \) — вектор предсказанных значений размерности \( m \times 1 \):</li>
</ul>

<div style="text-align:center;">
  $$ \hat{y} = \begin{bmatrix}
  \hat{y}^{(1)} \\
  \hat{y}^{(2)} \\
  \vdots \\
  \hat{y}^{(m)}
  \end{bmatrix} $$
</div>

<h2>Функция стоимости (ошибки)</h2>

<p>Целевая функция (или функция стоимости) — это функция, определяющая меру того, насколько хорошо модель предсказывает результаты. В контексте линейной регрессии, целевая функция измеряет разницу между предсказанными значениями и реальными значениями для всех объектов в наборе данных. Также данная функция называется функцией ошибки, потому что она показывает, насколько велика ошибка предсказаний.</p>

<p>Для оценки качества модели используем следующую функцию стоимости (ошибки)</p>

<div style="text-align:center;">
  $$ J(\theta) = \frac{1}{m} \sum_{i=1}^{m} \left( \hat{y}^{(i)} - y^{(i)} \right)^2 = \frac{1}{m} \| X\theta - y \|^2, $$
</div>

<p>где</p>
<ul>
  <li>\( y \) — вектор реальных значений целевой переменной,</li>
  <li>\( \hat{y} = X\theta \) — вектор предсказанных значений,</li>
  <li>\( \| \cdot \| \) — евклидова норма вектора.</li>
</ul>

<h3>Элементарная функция потерь для одного наблюдения</h3>

<div style="text-align:center;">
  $$ L^{(i)} = (y^{(i)} - \hat{y}^{(i)})^2 $$
</div>

<p>Данное выражение называют квадратом ошибки или квадратичным отклонением для \( i \)-го наблюдения. Функция стоимости \( J(\theta) \) является средним значением этой ошибки по всем наблюдениям, то есть средней квадратичной ошибкой (MSE).</p>

<blockquote>
<p>Функция потерь для одного наблюдения (описано выше) определяется как

<div style="text-align:center;">
  $$ L^{(i)} = (y^{(i)} - \hat{y}^{(i)})^2, $$
</div>

где $ {y}^{(i)} $ — истинное значение, а $ \hat{y}^{(i)} $ — предсказанное моделью. Это отклонение показывает, насколько сильно модель ошибается для каждого наблюдения.
</p>

<p>Средняя квадратичная ошибка (MSE) вычисляется как среднее этих квадратичных отклонений

<div style="text-align:center;">
  $$ \text{MSE} = \frac{1}{m} \sum_{i=1}^{m} (y^{(i)} - \hat{y}^{(i)})^2. $$
</div></p>

<p>MSE служит общей мерой качества модели на всем наборе данных.</p>

<p>Формула MSE похожа на формулу дисперсии случайной величины

<div style="text-align:center;">
  $$ \sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2, $$
</div>

где $ {x_i} $ — значение случайной величины, а $ \bar{x} $ — ее среднее значение.
</p>

<p>Обе формулы вычисляют среднее квадратов отклонений, но дисперсия измеряет вариацию значений вокруг среднего $ \bar{x} $ , тогда как MSE измеряет отклонение предсказанных значений $ \hat{y}^{(i)} $ от истинных значений $ {y}^{(i)} $</p>

<p>Таким образом, дисперсия анализирует вариацию вокруг одного среднего значения, тогда как MSE сравнивает <b>множество</b> предсказаний с <b>множеством</b> истинных значений.</p>
</blockquote>

<h3>Почему именно именно такая функция?</h3>

<p><ul>
  <li>Средний квадрат ошибки (MSE) позволяет оценить среднее отклонение предсказаний модели от реальных значений</li>
  <li>Квадрат ошибки способствует усилению влиянию больших отклонений, что немаловажно при минимазации крупных ошибок</li>
</ul></p>

<h2>Задача оптимизации</h2>

<p>Задача обучения модели линейной регрессии сводится к задаче оптимизации

<div style="text-align:center;">
  $$ \theta^* = \arg \min_{\theta} J(\theta), $$
</div>
где $ \theta^* $ — оптимальные параметры модели.
</p>

<h3>Почему это важно?</h3>

<p><ul>
  <li>Мы хотим определить такие значения параметров $ \theta $ , при которых модель будет наиболее точно предсказывать целевые значения на основе признаков</li>
  <li>Минимизацией функции ошибки мы находим параметры $ \theta $, которые лучше соответсвуют данным</li>
</ul></p>

<h3>Методы оптимизации</h3>

<p>После постановки задачи возникает вопрос: как найти параметры $ \theta $, минимизирующие функцию потерь $
J(\theta) $?</p>
<p>Существует несколько методов решения данной задачи. Рассмотрим некоторые из них.</p>

<h4>Метод наименьших квадратов (аналитическое решение)</h4>

<p>Метод наименьших квадратов предоставляет <b>аналитическое решение</b> для нахождения оптимальных параметров $ \theta $ в линейной регрессии. При определенных условиях (матрица $ X^T X $ обратима) возможно найти точное аналитическое решение. Для начала вычислим градиент функции стоимости по $ \theta $ :</p>

<p>1. Запишем функцию стоимости

<div style="text-align:center;"> $$ J(\theta) = \frac{1}{m} \sum_{i=1}^{m} (y^{(i)} - \hat{y}^{(i)})^2 = \frac{1}{m} \| y - X\theta \|^2, $$ </div></p>

<p>2. Функция стоимости выражается через квадрат евклидовой нормы

<div style="text-align:center;"> $$ J(\theta) = \frac{1}{m} (y - X\theta)^\top (y - X\theta). $$ </div></p>

<p>3. Градиент по $ \theta $ от функции стоимости $ J(\theta) $

<div style="text-align:center;"> $$ \nabla_\theta J(\theta) = \frac{1}{m} \frac{\partial}{\partial \theta} \left( (y - X\theta)^\top (y - X\theta) \right). $$ </div></p>

<p>4. Используя правило дифференцирования, получим

<div style="text-align:center;"> $$ \nabla_\theta J(\theta) = \frac{1}{m} \cdot 2 (X\theta - y)^\top \cdot \frac{\partial}{\partial \theta} (X\theta - y). $$ </div></p>

<p>5. Производная от линейной комбинации $ X \theta $ по $ \theta $ дает матрицу $ X $

<div style="text-align:center;">
  $$ \frac{\partial}{\partial \theta} (X\theta) = X. $$
</div></p>

<p>6. Таким образом, градиент принимает вид

<div style="text-align:center;">
  $$ \nabla_\theta J(\theta) = \frac{2}{m} (X\theta - y)^\top X. $$
</div></p>

<p>7. Транспонируем результат для получения конечной формы

<div style="text-align:center;"> $$ \nabla_\theta J(\theta) = \frac{2}{m} X^\top (X\theta - y). $$ </div>
</p>

<p>Теперь приравняем градиент к нулю для нахождения стационарных точек

<div style="text-align:center;">
  $$ \frac{2}{m} X^\top (X\theta - y) = 0. $$
</div></p>

<p>Найдем решение данного уравнения относительно $ \theta $

<div style="text-align:center;">
  $$ X^\top (X\theta - y) = 0, $$
  $$ \theta = (X^\top X)^{-1} X^\top y. $$
</div></p>

<h5>Почему найденную точку считаем минимумом?</h5>

<p><ul>
  <li>Функция $ J(\theta) $ выпуклая</li>
    <ul>
      <li>$ J(\theta) $ является квадратичной функцией относительно $ \theta $</li>
      <li>Матрица $ X^T X $ является положительно полуопределенной</li>
    </ul>
  <li>Выпуклость функции гарантирует, что любая стационарная точка является глобальным минимумом</li>
</ul></p>

<h5>Преимущества метода</h5>

<p><ul>
  <li>Возможность найти оптимальные параметры в одном вычислении</li>
  <li>Не требуется настройка гиперпараметров</li>
</ul></p>

<h5>Ограничения</h5>

<p><ul>
  <li>При большом числе признаков или наблюдений вычисление обратной матрицы может быть затратным</li>
  <li>Требование обратимости матрицы $ X^T X $</li>
</ul></p>

<h4>Метод градиентного спуска</h4>

<p>Градиентный спуск — итеративный алгоритм, позволяющий найти минимум функции потерь путем движения в направлении обратном градиенту, что хорошо себя показывает при больших наборах данных, где аналитическое решение затруднительно.</p>

<h4>Алгоритм градиентного спуска</h4>

<p>1. Инициализация параметров $ \theta^0 $</p>

<p>2. Для каждого шага $ k $
<ul>
  <li>Вычисляем предсказание
  <div style="text-align:center;">
    $$ \hat{y}^{(k)} = X \theta^{(k)}. $$
  </div></li>
  <li>Вычисляем градиент функции ошибки
  <div style="text-align:center;">
    $$ \nabla_\theta J(\theta^{(k)}) = \frac{2}{m} X^\top (\hat{y}^{(k)} - y). $$
  </div></li>
  <li>Обновляем параметры
  <div style="text-align:center;"> $$ \theta^{(k+1)} = \theta^{(k)} - \alpha \nabla_\theta J(\theta^{(k)}), $$ </div>
  где $ \alpha $ — скорость обучения.
  </li>
</ul></p>

<p>3. Повторяем до тех пор, пока изменение функции потерь не станет меньше заданного порога</p>

<h5>Преимущества метода</h5>

<p><ul>
  <li>Не требует вычисления обратной матрицы</li>
  <li>Позволяет использовать различные вариации (стохастический, пакетный и т.д)</li>
</ul></p>

<h5>Ограничения</h5>

<p><ul>
  <li>Требуется настройка гиперпараметра $ \alpha $</li>
  <li>Скорость сходимости может быть медленной</li>
</ul></p>

<h2>Математический аппарат</h2>

<p>Вектор — упорядоченный набор чисел, представляющий точку в многомерном пространстве.</p>

<p>Матрица — двумерный массив чисел, используемый для хранения и обработки данных.</p>

<p>Транспонирование — поворот матрицы относительно главной диагонали, обозначаемый $ A^\top $.</p>

<p>Обратная матрица —  матрица $ A^{-1} $ такая, что $ A A^{-1} = A^{-1} A = I $, где $ I $ — единичная матрица.</p>

<p>Скалярное произведение векторов для векторов $ a $ и $ b $
<div style="text-align:center;">
  $$ a^\top b = \sum_{i=1}^{n} a_i b_i. $$
</div></p>

<p>Евклидова норма (норма Фробениуса) векторов
<div style="text-align:center;">
 $$ \|a\| = \sqrt{a^\top a} = \sqrt{\sum_{i=1}^{n} a_i^2}. $$
</div></p>

<p>Для функции нескольких переменных $ f(\theta_0, \theta_1,...,\theta_n) $ частная производная по $ \theta_j $ определяется как
<div style="text-align:center;">
  $$ \frac{\partial f}{\partial \theta_j} = \lim_{\Delta \theta_j \to 0} \frac{f(\theta_0, \dots, \theta_j + \Delta \theta_j, \dots, \theta_n) - f(\theta_0, \dots, \theta_j, \dots, \theta_n)}{\Delta \theta_j} $$
</div></p>

<p>Градиент — это вектор, состоящий из частных производных по всем параметрам
<div style="text-align:center;">
  $$ \nabla_\theta f(\theta) = \begin{bmatrix} \frac{\partial f}{\partial \theta_0} \\ \frac{\partial f}{\partial \theta_1} \\ \vdots \\ \frac{\partial f}{\partial \theta_n} \end{bmatrix}. $$
</div></p>

<p>Правило дифференцирования квадратичной формы в общем виде
<div style="text-align:center;">
  $$ \frac{\partial}{\partial \theta} \left( v^\top A v \right) = 2 A v^\top \frac{\partial v}{\partial \theta}. $$
</div></p>











<h1>4. Практическая реализация линейной регрессии</h1>

<p>Реализуем линейную регрессию с нуля с помощью Python в середе Google Colab. Используем набор данных California Housing, который содержит информацию о ценах на жилье и связанные с ним характеристики.</p>

```python
"""загрузка необходимых библиотек"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
"""загрузка датасета"""

from sklearn.datasets import fetch_california_housing

# загрузка данных
california = fetch_california_housing()

# создание DataFrame
data = pd.DataFrame(california.data, columns=california.feature_names)
data['MedHouseValue'] = california.target

# просмотр первых 5 строк
data.head()
```

<p>Описание признаков:</p>
<ul>
  <li><strong>MedInc</strong> — медианный доход в район.</li>
  <li><strong>HouseAge</strong> — средний возраст домов в районе</li>
  <li><strong>AveRooms</strong> — среднее количество комнат на дом</li>
  <li><strong>AveBedrms</strong> — среднее количество спален на дом</li>
  <li><strong>Population</strong> — численность населения в районе</li>
  <li><strong>AveOccup</strong> — среднее количество жителей на дом</li>
  <li><strong>Latitude</strong> — широта района</li>
  <li><strong>Longitude</strong> — долгота района</li>
  <li><strong>MedHouseValue</strong> — медианная стоимость дома (целевой признак)</li>
</ul>


```python
"""основные статистические характеристики"""

data.describe()
```

```python
"""проверка пропущенных значений"""

data.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler

# Матрица признаков (размерность m x n)
X = data.drop('MedHouseValue', axis=1).values
# Вектор целевой переменной (размерность m x 1)
y = data['MedHouseValue'].values

# Нормализация признаков
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Количество наблюдений
m = len(y)
# добавление столбца единиц для свободного члена
X_b = np.hstack([np.ones((m, 1)), X_scaled])

# кол-во параметров (n признаков + 1)
n = X_b.shape[1]

# случайная инициализация параметров (размерность (n x 1))
theta_initial = np.random.randn(n, 1)
```

```python
# функция стоимости (ошибки)
def compute_cost(X, y, theta):
    m = len(y)
    y = y.reshape(-1, 1)
    predictions = X.dot(theta)
    errors = predictions - y
    cost = (1 / m) * np.sum(errors ** 2)
    return cost

# функция градиента
def compute_gradient(X, y, theta):
    m = len(y)
    y = y.reshape(-1, 1)
    errors = X.dot(theta) - y
    gradients = (2 / m) * X.T.dot(errors)
    return gradients

# алгоритм градиентного спуска
def gradient_descent(X, y, theta, learning_rate, num_iterations):
    cost_history = []
    for i in range(num_iterations):
        gradients = compute_gradient(X, y, theta)
        theta = theta - learning_rate * gradients
        cost = compute_cost(X, y, theta)
        cost_history.append(cost)
        if i % 50 == 0:
          print(f"Итерация {i} : Ошибка = {cost:.4f}")
    return theta, cost_history
```

```python
# обучение модели
learning_rate = 0.01
num_iterations = 2000

theta_initial = np.random.randn(n, 1) * 0.01
theta_optimal, cost_history = gradient_descent(X_b, y, theta_initial, learning_rate, num_iterations)
```

```python
"""визуализация функции потерь"""

plt.figure(figsize=(10, 6))
plt.plot(range(len(cost_history)), cost_history, 'b-')
plt.xlabel('Номер итерации')
plt.ylabel('Значение функции стоимости')
plt.title('Изменение функции стоимости во время обучения')
plt.show()
```

```python
# метрики качества
from sklearn.metrics import mean_squared_error, r2_score

# предсказания
y_pred = X_b.dot(theta_optimal)
# среднеквадратичная ошибка (MSE)
mse = mean_squared_error(y, y_pred)
print(f"Среднеквадратичная ошибка (MSE): {mse:.4f}")
# коэффициент детерминации R^2
r2 = r2_score(y, y_pred)
print(f"Коэффициент детерминации (R^2): {r2:.4f}")
```

```python
# реализация из коробки
from sklearn.linear_model import LinearRegression

# создание и обучение модели
model = LinearRegression()
model.fit(X_scaled, y)

# предсказания
y_pred_sklearn = model.predict(X_scaled)

# метрики
mse_sklearn = mean_squared_error(y, y_pred_sklearn)
r2_sklearn = r2_score(y, y_pred_sklearn)

print(f"\nScikit-learn MSE: {mse_sklearn:.4f}")
print(f"Scikit-learn R^2: {r2_sklearn:.4f}")
```

```python
"""Сравнение параметров модели"""

print("\nПараметры из реализации с нуля (градиентный спуск):")
for i in range(len(theta_optimal)):
    if i == 0:
        print(f"theta_0 (смещение): {theta_optimal[i][0]:.4f}")
    else:
        print(f"theta_{i} (коэффициент при признаке {data.columns[i-1]}): {theta_optimal[i][0]:.4f}")

print("\nПараметры из scikit-learn:")
print(f"theta_0 (смещение): {model.intercept_:.4f}")
for idx, coef in enumerate(model.coef_):
    print(f"theta_{idx+1} (коэффициент при признаке {data.columns[idx]}): {coef:.4f}")
```

```python
"""Сравнение метриик качества"""

print("\nСравнение метрик качества:")
print(f"Градиентный спуск MSE: {mse:.4f}, R^2: {r2:.4f}")
print(f"Scikit-learn MSE: {mse_sklearn:.4f}, R^2: {r2_sklearn:.4f}")
```

```python
"""Вычисление параметров с помощью аналитического решения"""

theta_normal = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y.reshape(-1, 1))
```

```python
"""Сравнение параметров"""

print("\nПараметры из аналитического решения (аналитическое решение):")
for i in range(len(theta_normal)):
    if i == 0:
        print(f"theta_0 (смещение): {theta_normal[i][0]:.4f}")
    else:
        print(f"theta_{i} (коэффициент при признаке {data.columns[i-1]}): {theta_normal[i][0]:.4f}")
```

```python
"""Вычисление метрик аналитического решения"""

# предсказания
y_pred_normal = X_b.dot(theta_normal)

# метрики
mse_normal = mean_squared_error(y, y_pred_normal)
r2_normal = r2_score(y, y_pred_normal)

print(f"\nАналитическое решение MSE: {mse_normal:.4f}")
print(f"Аналитическое решение R^2: {r2_normal:.4f}")
```

```python
"""Сравнение качества всех моделей"""

print(f"Градиентный спуск MSE: {mse:.4f}, R^2: {r2:.4f}")
print(f"Scikit-learn MSE: {mse_sklearn:.4f}, R^2: {r2_sklearn:.4f}")
print(f"Аналитическое решение MSE: {mse_normal:.4f}, R^2: {r2_normal:.4f}")
```
